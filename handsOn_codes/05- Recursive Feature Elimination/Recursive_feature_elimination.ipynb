```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
import matplotlib.pyplot as plt

# Load the MNIST digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a logistic regression model (you can choose a different model as well)
model = LogisticRegression(max_iter=10000)

# Create an RFE selector with cross-validation
rfe = RFE(estimator=model, n_features_to_select=1, step=1)

# Fit the RFE selector on the training data
rfe.fit(X_train, y_train)

# Get the ranking of each pixel
pixel_ranking = rfe.ranking_

# Plot the pixel ranking to visualize feature importance
plt.figure(figsize=(8, 6))
plt.imshow(pixel_ranking.reshape(8, 8), cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title("Pixel Ranking (1=Most Important)")
plt.show()
```
