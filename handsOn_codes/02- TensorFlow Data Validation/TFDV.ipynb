
### Import TensorFlow and other packages


```python
import tensorflow_data_validation as tfdv
import pandas as pd
import tensorflow as tf
import tensorflow_data_validation as tfdv
print('TF version:', tf.__version__)
print('TFDV version:', tfdv.version.__version__)
```

## Load the dataset


```python
# Load your dataset (this is just a simple example, replace with your own data)
data = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

##Computing descriptive data statistics

```python
# Generate statistics for the dataset
stats = tfdv.generate_statistics_from_dataframe(data)
```

## Visualize the statistics


```python
# Visualize the statistics
tfdv.visualize_statistics(stats)
```

## Inferring a schema over the data

```python
# Infer a schema from the statistics
schema = tfdv.infer_schema(stats)
```

```python
# schema
```

```python
tfdv.display_schema(schema)
```

## Check evaluation data for errors

```python
#Validate the dataset against the inferred schema
anomalies = tfdv.validate_statistics(stats, schema)

# Print detected anomalies
# print(anomalies)
```

```python
tfdv.display_anomalies(anomalies)
```
